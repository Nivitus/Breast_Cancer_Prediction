```python
X = df.iloc[:,2:31].values
Y = df.iloc[:,1].values
```

```python
# Spliting the data set 75% Training and 25% Testing dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
# scale the data(Feature Scaling)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc =StandardScaler()
sc.fit_transform(X_train)
sc.fit_transform(X_test)
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Create the function for models
```

```python
def models(X_train,X_test):
    
    # Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    
    # Decision tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(random_state=0,criterion='entropy')
    tree.fit(X_train,Y_train)
    
    # Random Forest
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators=10,random_state=0,criterion='entropy')
    forest.fit(X_train,Y_train)
    
    # Print the models accuracy
    print('(1) Logistic Training Accuracy:', log.score(X_train,Y_train))
    print('(2) Decision Tree:', tree.score(X_train,Y_train))
    print('(2) Random Forest Training Accuracy:', forest.score(X_train,Y_train))
    
    return log, tree, forest
```

```python
model = models(X_train,Y_train)
```

```python
# test mmodel accuracy finding using confusion matrix
```

```python
from sklearn.metrics import confusion_matrix

for i in range(len(model)):
    print('Model', i)
    cm = confusion_matrix(Y_test,model[i].predict(X_test))

    TP = cm[0][0]
    TN = cm[1][1]
    FN = cm[1][0]
    FP = cm[0][1]

    print(cm)
    print('Tesing Accuracy = ',(TP + TN) / (TP + TN + FP + FN))
    print()
```

```python
# Another way to find accuracy metrics
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
for i in range(len(model)):
    print('Model', i)

    print(classification_report(Y_test, model[i].predict(X_test)))
    print(accuracy_score(Y_test, model[i].predict(X_test)))
```
